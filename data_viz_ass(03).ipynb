```python
#swarmplot
titanic=snr.load_dataset('titanic')
snr.swarmplot(data=titanic,x='fare')
```

```python
snr.swarmplot(data=titanic,x='fare',y='age')
```

```python
snr.swarmplot(data=titanic,x='pclass',y='age')
```

majority of 3rd class people of aged 15-35 contributed for more fares

```python
snr.swarmplot(data=titanic,x='sex',y='age')
```

majority of male under age 20-35 have contributed more fares

```python
snr.swarmplot(data=titanic,x='sex',y='age',hue='pclass',dodge=True)
```

```python
snr.swarmplot(data=titanic,x='sex',y='age',orient='h',hue='pclass',dodge=True)
```
